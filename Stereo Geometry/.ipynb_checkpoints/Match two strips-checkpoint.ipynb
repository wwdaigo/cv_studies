# Compute the disparity between two strips

```python
import numpy as np
import cv2
from matplotlib import pyplot
```

### Load the images

```python
left = cv2.imread("input/flowers-left.png")
right = cv2.imread("input/flowers-right.png")

%pylab inline
pylab.rcParams['figure.figsize'] = (20, 10)

plt.figure(frameon=False)
fig, (imgL, imgR) = plt.subplots(1, 2)
imgL.imshow(left)
imgR.imshow(right)
```

### Convert to grayscale

```python
gray_left = cv2.cvtColor(left, cv2.COLOR_BGR2GRAY)
gray_right = cv2.cvtColor(right, cv2.COLOR_BGR2GRAY)

fig, (imgL, imgR) = plt.subplots(1, 2)
imgL.imshow(gray_left, cmap = plt.cm.binary)
imgR.imshow(gray_right, cmap = plt.cm.binary)
```

### Define the strip row and the block size

```python
y=120
b=100
```

### Extract the strips

```python
strip_left = gray_left[y:y+b, :]
strip_right = gray_right[y:y+b, :]

%pylab inline
pylab.rcParams['figure.figsize'] = (10, 4)

plt.figure(frameon=False)
fig, (imgL, imgR) = plt.subplots(2, 1)
imgL.imshow(strip_left, cmap=plt.cm.binary)
imgR.imshow(strip_right, cmap=plt.cm.binary)
```

### Disparity function

```python
def best_x(patch, strip):
    min_diff = np.inf
    best_x = -1
    
    patch_size = patch.shape
    number_of_cols = strip.shape[1] - patch.shape[1]
    for i in range(number_of_cols):
        other_patch = strip[:, i:i+patch_size[1]]
        diff = np.sum((patch-other_patch)**2)
        
        if diff < min_diff:
            min_diff = diff
            best_x = i
    
    return best_x
```

```python
def disparity(s_left, s_right, block_size):
    width = s_left.shape[1]
    number_of_columns = width // block_size

    disparities = np.zeros((number_of_columns, 1))
    
    for i in range(number_of_columns):
        x_left = i * block_size
        block_left = s_left[:, x_left:x_left+block_size]
        x_right = best_x(block_left, s_right)
        block_right = s_right[:, x_left:x_left+block_size]
    
        disparities[i] = x_left - x_right
    
    return disparities
```

```python
disps = disparity(strip_left, strip_right, 80)
plt.plot(disps)
disps
```
