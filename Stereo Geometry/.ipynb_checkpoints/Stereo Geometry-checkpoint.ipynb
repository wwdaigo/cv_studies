## Question 1

Use the pair pair0-L.png and pair0-R.png - a central square moved 2 pixels horizontally:

Implement the basic stereo algorithm of taking a window around every pixel in one image and search for the best match along the same scan line in the other image. You will do this both left to right and right to left.  Remember: Because of depth changes (discontinuities) some pixels visible in the left image are not in the right and vice a versa. So you will match in both directions.
For this part implement the simplest thing imaginable: Look for the smallest difference between the template window (source) and the proposed location window. Use the sum of squared differences measure (SSD). We are going to take the definitions from: https://software.intel.com/en-us/node/504333
SSD is defined by:

$$S_{tx}(r,c)=\sum_{j=0}^{tplRows-1}\sum_{i=0}^{tplCols-1}\left[t(j,i)-x\left(r+j-\frac{tplRows}{2}, c+i-\frac{tplCols}{2}\right)\right]^{2}$$

Basically you just sum up the squares.  A “good” match, then, is when this value is at a minimum. That is, you are looking for the same image patch in both images.
Implement the SSD match algorithm as function disparity_ssd(L, R) that returns a disparity image D(y,x) such that L(y,x) = R(y,x+D(y,x)) when matching from left (L) to right (R).

### Extract the patch from image

*Input* image, coordinates and blocksize  
*Output* The image matrix

````python
def create_coords(central_coord, block_size):
    x = central_coord[0]
    y = central_coord[1]
    
    padding = (block_size) // 2
    
    left_x = max(x - padding, 0)
    top_y = max(y - padding, 0)
    
    right_x = x + padding + 1
    bottom_y = y + padding + 1
    
    return left_x, top_y, right_x, bottom_y
````

````python
def extract_patch(image_matrix, central_coord, block_size):
    left_x, top_y, right_x, bottom_y = create_coords(central_coord, block_size)
    
    return image_matrix[top_y:bottom_y, left_x:right_x]
````

#### Testing the patch extraction

### Extract the strip

````python
def extract_strip(image, central_coord, block_size):
    _, top_y, _, bottom_y = create_coords(central_coord, block_size)
    return image[top_y:bottom_y, :]
````

````python
strip_left_image_test = extract_strip(left_image, (100, 128), 5)

plt.imshow(strip_left_image_test)
````

### Calculate the best match X

````python
def best_match_x(patch, strip):
    min_diff = np.inf
    best_x = -1
    
    patch_size = patch.shape
    number_of_cols = strip.shape[1] - patch.shape[1]

    for i in range(number_of_cols):
        other_patch = strip[:, i:i+patch_size[1]]
        diff = np.sum((patch-other_patch)**2)
        
        if diff < min_diff:
            min_diff = diff
            best_x = i
    
    return best_x
````

### Normalize the output to 0-1 range

````python
def normalize(matrix):
    return (matrix-np.min(matrix))/(np.max(matrix)-np.min(matrix))
````

### Calculate the disparity

````python
def disparity_ssd(left, right):
    """Compute disparity map D(y, x) such that: L(y, x) = R(y, x + D(y, x))
    
    Params:
    L: Grayscale left image (Source)
    R: Grayscale right image, same size as L (Template)

    Returns: Disparity map, same size as L, R
    """
    
    # TODO: Your code here
    width = left.shape[1]
    height = left.shape[0]
    
    block_size = 5
    
    disparities = np.zeros((height, width))
    
    for y in range(height):
        for x in range(width):
            patch_left = extract_patch(left, (x, y), block_size)
            strip_right = extract_strip(right, (x, y), block_size)
            best_x = best_match_x(patch_left, strip_right)
            disparities[y][x] = x - best_x
    
    return disparities
````

````python
D_L = disparity_ssd(left_image, right_image)
L_D = disparity_ssd(right_image, left_image)
````

````python
def image_disparity_output(image, disparity):
    return image + disparity
````

````python
out_D_L = image_disparity_output(right_image, D_L)
plt.imshow(normalize(out_D_L), cmap=plt.cm.binary)
````

````python
out_L_D = image_disparity_output(left_image, L_D)
plt.imshow(normalize(out_L_D), cmap=plt.cm.binary)
````
